```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
```

```python
data = pd.read_csv('data untuk dicleansing.csv', sep=";")
```

```python
data = pd.read_csv('data untuk dicleansing.csv', sep=";")
while True:
    # Langkah 2: Melakukan pengelompokan berdasarkan kolom "Material"
    outlier_columns = []
    grouped = data.groupby('Material')
    cleaned_df = pd.DataFrame()
    for material, group in grouped:
        Q1 = group['Qty'].quantile(0.25)
        Q3 = group['Qty'].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers = group[(group['Qty'] < lower_fence) | (group['Qty'] > upper_fence)]
        outliers['Qty_Outliers'] = True
        non_outliers = group[(group['Qty'] >= lower_fence) & (group['Qty'] <= upper_fence)]
        non_outliers['Qty_Outliers'] = False
        cleaned_df = pd.concat([cleaned_df, outliers, non_outliers])
    outlier_columns.append("Qty_Outliers")
    data = cleaned_df
    grouped = data.groupby('Material')
    cleaned_df = pd.DataFrame()
    for material, group in grouped:
        Q1 = group['pr-po'].quantile(0.25)
        Q3 = group['pr-po'].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers = group[(group['pr-po'] < lower_fence) | (group['pr-po'] > upper_fence)]
        outliers['pr-po_Outliers'] = True
        non_outliers = group[(group['pr-po'] >= lower_fence) & (group['pr-po'] <= upper_fence)]
        non_outliers['pr-po_Outliers'] = False
        cleaned_df = pd.concat([cleaned_df, outliers, non_outliers])
    outlier_columns.append("pr-po_Outliers")
    data = cleaned_df
    grouped = data.groupby('Nama Vendor')
    cleaned_df = pd.DataFrame()
    for material, group in grouped:
        Q1 = group['po-del'].quantile(0.25)
        Q3 = group['po-del'].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers = group[(group['po-del'] < lower_fence) | (group['po-del'] > upper_fence)]
        outliers['po-del_Outliers'] = True
        non_outliers = group[(group['po-del'] >= lower_fence) & (group['po-del'] <= upper_fence)]
        non_outliers['po-del_Outliers'] = False
        cleaned_df = pd.concat([cleaned_df, outliers, non_outliers])
    outlier_columns.append("po-del_Outliers")
    data = cleaned_df
    grouped = data.groupby('Nama Vendor')
    cleaned_df = pd.DataFrame()
    for material, group in grouped:
        Q1 = group['po-gr'].quantile(0.25)
        Q3 = group['po-gr'].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers = group[(group['po-gr'] < lower_fence) | (group['po-gr'] > upper_fence)]
        outliers['po-gr_Outliers'] = True
        non_outliers = group[(group['po-gr'] >= lower_fence) & (group['po-gr'] <= upper_fence)]
        non_outliers['po-gr_Outliers'] = False
        cleaned_df = pd.concat([cleaned_df, outliers, non_outliers])
    outlier_columns.append("po-gr_Outliers")
    data = cleaned_df
    cleaned_df = pd.DataFrame()
    grouped = data.groupby('Nama Vendor')
    for material, group in grouped:
        Q1 = group['selisih'].quantile(0.25)
        Q3 = group['selisih'].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        outliers = group[(group['selisih'] < lower_fence) | (group['selisih'] > upper_fence)]
        outliers['selisih_Outliers'] = True
        non_outliers = group[(group['selisih'] >= lower_fence) & (group['selisih'] <= upper_fence)]
        non_outliers['selisih_Outliers'] = False
        cleaned_df = pd.concat([cleaned_df, outliers, non_outliers])
    outlier_columns.append("selisih_Outliers")
    data = cleaned_df
    data['num_outliers'] = data[outlier_columns].sum(axis=1)
    # Langkah 4: Buang row data yang memiliki nilai "Banyak_kolom_outlier" tertinggi
    max_outliers_row = data['num_outliers'].max()
    if max_outliers_row == 0:
        data = data.drop(columns=outlier_columns)
        data = data.drop(columns=['num_outliers'])
        break
    data = data[data['num_outliers'] < max_outliers_row]
    data = data.drop(columns=outlier_columns)
    data = data.drop(columns=['num_outliers'])    
# Langkah 6: Menyimpan data tanpa outlier ke dalam file Excel
material_freq = data['Material'].value_counts()
vendor_freq = data['Nama Vendor'].value_counts()
material_freq = material_freq[material_freq >= 10]
vendor_freq = vendor_freq[vendor_freq >= 10]
data = data[data['Material'].isin(material_freq.index) & data['Nama Vendor'].isin(vendor_freq.index)]
output_file = 'bersihin outlier + filter frekuensi v1 baru.xlsx'  # Ganti dengan nama file Excel tujuan Anda
data.to_excel(output_file, index=False)
print(f"Data tanpa outlier telah disimpan ke dalam file {output_file}")
```
